```python
import pandas as pd
from sentence_transformers import SentenceTransformer
import numpy as np
from sklearn.metrics.pairwise import cosine_similarity

# Load your comments data
df = pd.read_csv('comments.csv')
```

```python
# Load a pretrained Sentence Transformer model
model = SentenceTransformer('sentence-transformers/all-mpnet-base-v2')
```

```python
# Define predefined sentiment examples (you can adjust these)
positive_example = "This is great, I love it!"
negative_example = "This is terrible, I hate it!"
neutral_example = "This is just okay, nothing special."
```

```python

# Create embeddings for sentiment examples
positive_embedding = model.encode(positive_example)
negative_embedding = model.encode(negative_example)
neutral_embedding = model.encode(neutral_example)
```

```python
# Function to classify sentiment based on similarity
def classify_sentiment(comment):
    # Get embedding for the comment
    comment_embedding = model.encode(comment)
    
    # Compute cosine similarities between the comment and predefined examples
    positive_similarity = cosine_similarity([comment_embedding], [positive_embedding])[0][0]
    negative_similarity = cosine_similarity([comment_embedding], [negative_embedding])[0][0]
    neutral_similarity = cosine_similarity([comment_embedding], [neutral_embedding])[0][0]
    
    # Classify based on the highest similarity
    similarities = [positive_similarity, negative_similarity, neutral_similarity]
    sentiment_labels = ['positive', 'negative', 'neutral']
    
    # Return the sentiment label with the highest similarity
    return sentiment_labels[np.argmax(similarities)]
```

```python
# Apply the sentiment classification to the 'commentText' column
df['sentiment'] = df['commentText'].apply(classify_sentiment)
```

```python

# Save the labeled data (optional)
df.to_csv('labeled_comments.csv', index=False)
```
